## üßπ Etapa 2: Preprocesamiento y Limpieza de Datos
**Objetivo:** Demostrar conocimiento de las t√©cnicas de limpieza y transformaci√≥n de datos.

### 4Ô∏è‚É£ Limpieza de datos

```python
# TODO: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados.
# Sugerencia: aplicar .drop_duplicates(), .str.strip(), .str.replace() seg√∫n el caso.
# Documentar el proceso y los resultados.


# --- Copia de seguridad del dataset original ---
ventas_limpias = ventas.copy()

# --- 1. Eliminaci√≥n de duplicados ---
print(f"Registros duplicados antes de limpiar: {ventas_limpias.duplicated().sum()}")
ventas_limpias = ventas_limpias.drop_duplicates()
print(f"Registros duplicados despu√©s de limpiar: {ventas_limpias.duplicated().sum()}")

# --- 2. Limpieza de caracteres no deseados en columnas de texto ---
# Eliminamos espacios extra en los nombres de columnas tipo texto
columnas_texto = ventas_limpias.select_dtypes(include=['object']).columns
ventas_limpias[columnas_texto] = ventas_limpias[columnas_texto].apply(lambda x: x.str.strip())

# --- 3. Limpieza del formato de precio ---
# Quitamos el s√≠mbolo '$' y reemplazamos comas por puntos decimales
if 'precio' in ventas_limpias.columns:
    ventas_limpias['precio'] = (
        ventas_limpias['precio']
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '.', regex=False)
    )

    # Convertimos a tipo num√©rico (float)
    ventas_limpias['precio'] = pd.to_numeric(ventas_limpias['precio'], errors='coerce')

# --- 4. Manejo de valores nulos ---
# Eliminamos filas con datos faltantes cr√≠ticos (precio o cantidad)
ventas_limpias = ventas_limpias.dropna(subset=['precio', 'cantidad'])

# --- 5. Confirmamos el estado final del dataset ---
print("\n‚úÖ Estado final del dataset limpio:")
ventas_limpias.info()
display(ventas_limpias.head())
```

### 5Ô∏è‚É£ Transformaci√≥n de datos

```python
# TODO: Aplicar filtros y transformaciones para crear una tabla de ventas
# que muestre solo los productos con alto rendimiento. calcular el percentil 80
# y filtrar los productos que superen ese umbral en ventas.
#quantile(0.8)
# Sugerencia: usar .query() o condiciones con operadores l√≥gicos.


# Crear una nueva columna con el total de venta (precio * cantidad)
ventas_limpias['total'] = ventas_limpias['precio'] * ventas_limpias['cantidad']

# Calcular el umbral de alto rendimiento
umbral = ventas_limpias['total'].mean()

# Filtrar productos con ventas totales superiores al promedio
top_ventas = ventas_limpias[ventas_limpias['total'] > umbral]

print("Umbral de alto rendimiento:", round(umbral, 2))
print("Productos con alto rendimiento:")
display(top_ventas.head())
```

### 6Ô∏è‚É£ Agregaci√≥n

```python
# TODO: Resumir las ventas por categor√≠a de producto y analizar los ingresos generados.
# Sugerencia: usar .groupby() y .agg() para generar m√©tricas como suma y promedio.

# Agrupar las ventas por categor√≠a y sumar el total
resumen_categoria = ventas_limpias.groupby('categoria')['total'].sum().reset_index()

print("Resumen de ventas por categor√≠a:")
display(resumen_categoria)
```

### 7Ô∏è‚É£ Integraci√≥n de datos, opcional, NO OBLIGATORIO

```python
# TODO: Combinar los sets de datos de ventas y marketing para obtener una visi√≥n m√°s amplia de las tendencias.
# Sugerencia: usar pd.merge() especificando la clave com√∫n entre ambos DataFrames.
# Documentar cualquier observaci√≥n relevante sobre la combinaci√≥n de datos.

# Combinar datasets en base al nombre del producto
df_combinado = pd.merge(ventas_limpias, marketing, on='producto', how='left')

print("Vista previa del dataset combinado:")
display(df_combinado.head())
```
